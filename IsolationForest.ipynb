```python
import h2o as h
import pandas as pd
h.init()
```

```python
df = pd.read_csv('train_v1.csv')
test_df = pd.read_csv('test_v1.csv')
```

```python
train_hf = h.H2OFrame(df)
test_hf = h.H2OFrame(test_df)
```

```python
features = train_hf.col_names
features.remove("Anomaly")
features.remove('Index')
```

```python
features
```

```python
from h2o.estimators import H2OIsolationForestEstimator

if_model = H2OIsolationForestEstimator(
    seed=42,
    ntrees=100,
    max_depth=10,
    sample_size=500
)

if_model.train(training_frame=train_hf[features])
```

```python
anomaly_scores = if_model.predict(train_hf)
```

```python
anomaly_scores_df = h.as_list(anomaly_scores)
```

```python
treshhold = 0.5 # for example
df['anom_pred'] = (anomaly_scores_df['predict'] >= treshhold).astype(int)
```

```python
from sklearn.metrics import classification_report
report = classification_report(df['Anomaly'],df['anom_pred'], output_dict=True)
```

```python
precision = report['1']['precision']
recall = report['1']['recall']
f1_score = report['1']['f1-score']

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1_score:.2f}")
```

```python
from h2o.estimators import H2OIsolationForestEstimator
from h2o.grid import H2OGridSearch
hyper_params = {
    'ntrees': [50, 100, 150, 200],
    'max_depth': [5, 10, 15, 20],
    'sample_size': [256, 500, 750, 1000]
}
grid_search = H2OGridSearch(
    model=H2OIsolationForestEstimator(seed=42),
    hyper_params=hyper_params,
    search_criteria={
        'strategy': 'RandomDiscrete', 
        'max_models': 50, 
        'stopping_metric': 'deviance', 
        'stopping_tolerance': 0.001,
        'stopping_rounds': 5
    }
)
```

```python
grid_search.train(training_frame=train_hf[features])
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

def evaluate_model(model, threshold_range, data, true_labels):
    # Calculate anomaly scores
    anomaly_scores = model.predict(data)
    anomaly_scores_df = h.as_list(anomaly_scores)
    
    best_f1 = -1
    best_threshold = None
    best_precision = None
    best_recall = None
    
    for threshold in threshold_range:
        # Get predictions based on the threshold
        predictions = (anomaly_scores_df["predict"] >= threshold).astype(int)
        
        # Calculate precision, recall, and F1 score
        precision = precision_score(true_labels, predictions)
        recall = recall_score(true_labels, predictions)
        f1 = f1_score(true_labels, predictions)
        
        # Update the best values if the current F1 score is better
        if f1 > best_f1:
            best_f1 = f1
            best_threshold = threshold
            best_precision = precision
            best_recall = recall
            
    return best_threshold, best_precision, best_recall, best_f1
```

```python
import numpy as np
# Convert the H2OFrame to a Pandas DataFrame
data_df = train_hf.as_data_frame()

# Get the true labels
true_labels = data_df["Anomaly"]

# Test different threshold values
threshold_range = np.linspace(0, 1, 50)

# Evaluate all the models and choose the best one
best_model = None
best_threshold = None
best_precision = None
best_recall = None
best_f1 = -1

for model in grid_search.models:
    threshold, precision, recall, f1 = evaluate_model(model, threshold_range, train_hf, true_labels)
    if f1 > best_f1:
        best_model = model
        best_threshold = threshold
        best_precision = precision
        best_recall = recall
        best_f1 = f1

print("Best Model:", best_model.model_id)
print("Best Threshold:", best_threshold)
print("Precision:", best_precision)
print("Recall:", best_recall)
print("F1 Score:", best_f1)
```

```python
test_anomaly_scores = best_model.predict(test_hf)
test_anomaly_scores_df = h.as_list(test_anomaly_scores)
```

```python
test_df["Anom_Pred"] = (test_anomaly_scores_df["predict"] >= best_threshold).astype(int)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

test_actual_labels = test_df["Anomaly"].values
test_predicted_labels = test_df["Anom_Pred"].values

test_precision = precision_score(test_actual_labels, test_predicted_labels)
test_recall = recall_score(test_actual_labels, test_predicted_labels)
test_f1 = f1_score(test_actual_labels, test_predicted_labels)

print("Test Precision:", test_precision)
print("Test Recall:", test_recall)
print("Test F1 Score:", test_f1)
```
